```python
df = df.filter(items=['Desire-Expression-Sentence','Prior-Context','Post-Context','Fulfillment-Label'])
```

```python
df = df.loc[df['Fulfillment-Label'].isin(['Fulfilled', 'Unfulfilled', 'Unknown from the context'])]
```

```python
df.head(20)
```

```python
prio = df['Prior-Context'].values
post = df['Post-Context'].values
proof = df['Desire-Expression-Sentence'].values
labels = df['Fulfillment-Label'].values
```

```python
labels = ['Unknown' if x == 'Unknown from the context' else x for x in labels]
```

```python
content = [x for x in proof]
```

```python
data_build = {'Content' : content, 'Label' : labels}

new_df = pd.DataFrame(data_build)
```
